```python
import pandas as pd
from consts import N_HOTELS
import pickle
import numpy as np
```

```python
def convert_string_to_nparray(s):
    return np.array([int(x) for x in s.strip('[]').split(', ')])
```

# Load Data from 2_create_reviews_categories_vectors.ipynb

```python
df=pd.read_csv('/home/student/project/HumanChoicePrediction/data/df_per_hotel_reviews_scores.csv')
df.head()
```

# Create scores dict

```python
result = {}

for column in df[['ID', 'positive_to_negative_sentiment_ratio', 'positive_to_negative_length_ratio', 'hotel_metrices_vector']].columns:
    if column != 'ID':
        if column == 'hotel_metrices_vector':
            result[column] = df.set_index('ID')[column].apply(convert_string_to_nparray).to_dict()
        else:

            result[column] = df.set_index('ID')[column].to_dict()

display(result)
```

# Save dict as pickle

```python
with open('/home/student/project/HumanChoicePrediction/data/data_scores_dict.pkl', 'wb') as f:
    pickle.dump(result, f)
```
